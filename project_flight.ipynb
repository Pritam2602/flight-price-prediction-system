```python
def dur(x):
    parts = x.split(' ')
    hours = '0h'
    minutes = '0m'
    for i in parts:
        if 'h' in i:
            hours = i
        elif 'm' in i:
            minutes = i

    return f"{hours} {minutes}"
            


        
```

```python
Data['Duration'] = Data['Duration'].apply(dur)
```

```python
Data['Duration_hour'] = Data['Duration'].apply(lambda x:int(x.split(' ')[0][0:-1]))
```

```python
Data['Duration_minute'] = Data['Duration'].apply(lambda x:int(x.split(' ')[1][0:-1]))
```

```python
def time(x):
    if x>4 and x<=8:
        return "Early Morning"
    elif x>8 and x<=12:
        return "Morning"
    elif x>12 and x<=16:
        return "Noon"
    elif x>16 and x<=20:
        return "Evening"
    elif x>20 and x<=24:
        return "Night"
    else:
        return "Late Night"
        
    
```

```python
q3 = Data["Price"].quantile(0.75)
q1 = Data["Price"].quantile(0.25)
iqr = q3 - q1
max_val = q3 + 1.5*iqr
min_val = q1 - 1.5*iqr
Data["price"] = np.where(Data["Price"]>max_val ,Data["Price"].median(),Data["Price"])
Data.head()
```

```python
Data["Journey_year"]
x = Data.drop("price",axis=1)
y = Data["Price"]
x = x.drop("Price",axis=1)
x
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
for train_index,test_index in kf.split(x):
    x_train,x_test = x.iloc[train_index],x.iloc[test_index]
    y_train,y_test = y.iloc[train_index],y.iloc[test_index]
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
!pip install xgboost
```

```python
import xgboost as xgb
```

```python
xgb_model = xgb.XGBRegressor()
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=kf, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(x_train,y_train)
```

```python

best_param = grid_search.best_params_
```

```python
best_param
```

```python
xgb_best = xgb.XGBRegressor(**best_param)
```

```python
xgb_best.fit(x_train, y_train)
```

```python
y_predict = xgb_best.predict(x_test)
```

```python
residuals = y_test - y_predict
```

```python
print("Residuals mean:", np.mean(residuals))
```

```python
# Plotting residuals
plt.figure(figsize=(10, 5))
sns.histplot(residuals, kde=True)
plt.axvline(x=0, color='r', linestyle='--')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Residuals Distribution')
plt.show()
```
